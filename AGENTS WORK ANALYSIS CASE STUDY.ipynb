# STRUCTURE
An incentive structure for agents in the form of a per task rate and tasks completed.

```python
#minimum agent per task and the number of agents
min_payment = 175
no_agents = 1199

np.random.seed(42)

#Random completion
tasks_completed = np.random.poisson(10, no_agents)
df3 = pd.DataFrame({'agent_id': range(1, no_agents +1 ), 'tasks_completed': tasks_completed})

#Calculate quartiles
Q1 = df3['tasks_completed'].quantile(0.25)
Q2 = df3['tasks_completed'].quantile(0.50)
Q3 = df3['tasks_completed'].quantile(0.75)

#Assign quartiles and calculate per_taks rates
def calculate_bonus(row):
    if row['tasks_completed'] < Q1:
        return min_payment
    elif Q1 <= row['tasks_completed'] < Q2:
        return min_payment + 10 #increase for Q2
    elif Q1 <= row['tasks_completed'] < Q3:
        return min_payment + 20 #increase for Q3
    else:
        return min_payment + 30 #increase for Q4

df3['per_task_rate'] = df3.apply(calculate_bonus, axis = 1)

print(df3[['agent_id','tasks_completed','per_task_rate']])
```
